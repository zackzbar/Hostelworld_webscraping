```python
import pandas as pd
import matplotlib.pyplot as plt

full = pd.read_csv('full.csv', index_col=0)

pd.set_option('display.max_rows', 999)
pd.set_option('display.min_rows', 999)
```

```python
mains = ['name','kind','city','country','price','reviews','rating_cat','rating','value']
```

```python
# COUNTING THE FULL DATASET
```

```python
full_countrycount = pd.DataFrame(full[full['reviews']>=25]['country'].value_counts())
full_countrycount
```

```python
full_countrycount.plot.pie(y='country',figsize=(10,10),legend=False,autopct='%1.0f%%')#,textprops={'color':"w"})
plt.show()
```

```python
# MEASURES OF CENTRALITY FOR EACH COUNTRY, VARIOUS COLUMNS
```

```python
full[full['reviews']>25].groupby('country')['price'].agg(['mean','median']).sort_values(by='median',ascending=False)
```

```python
# VALUE CALCULATION AND RATING
```

```python
value_calc = full[(full['reviews']>=25) & (full['value']>0)].sort_values('value', ascending=False)
value_rate = full[(full['reviews']>=25) & (full['value']>0)].sort_values('value_for_money', ascending=False)
value_calc.shape
```

```python
display(value_calc[mains].head(100))
value_calc[mains].tail(10)
```

```python
mains.append('value_for_money')

display(value_rate[mains].head(100))
value_rate[mains].tail(10)
```
