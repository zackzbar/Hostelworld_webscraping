```python
# CORRELATIONS!!
```

```python
list(value_calc)[-1:-10:-1] 
```

```python
from scipy.stats import pearsonr

df = pd.DataFrame(columns=['feature','corr_reviews','corr_rating'])
value_calc_corr = value_calc[value_calc['reviews']<5000]

for x in list(value_calc)[-1:-10:-1] :
    calc, _ = pearsonr(value_calc_corr[x], value_calc_corr['reviews'])
    calc2, _ = pearsonr(value_calc_corr[x], value_calc_corr['rating'])
    df = df.append({'feature':x, 'corr_reviews':calc, 'corr_rating':calc2}, ignore_index=True)

print('By reviews:')
display(df.sort_values(by=['corr_reviews'], ascending=False))
print('By rating:')
display(df.sort_values(by=['corr_rating'], ascending=False))
```

```python
df[['feature','corr_reviews']].set_index('feature').sort_values(by=['corr_reviews'], ascending=False)[:-1].plot.bar(legend=False)
plt.show()

value_calc.plot(kind='scatter',x='atmosphere',y='reviews')
plt.ylim(0,500)
```

```python
trio = ['Vietnam','Laos','Cambodia']

df_trio = value_calc[value_calc['country'].isin(trio)]
```

```python
df = pd.DataFrame(columns=['feature','corr_reviews','corr_rating'])
df_trio_corr = df_trio[df_trio['reviews']<5000]

for x in list(df_trio)[-1:-10:-1] :
    calc, _ = pearsonr(df_trio_corr[x], df_trio_corr['reviews'])
    calc2, _ = pearsonr(df_trio_corr[x], df_trio_corr['rating'])
    df = df.append({'feature':x, 'corr_reviews':calc, 'corr_rating':calc2}, ignore_index=True)

print('Sorted by reviews:')
display(df.sort_values(by=['corr_reviews'], ascending=False))
print('Sorted by rating:')
display(df.sort_values(by=['corr_rating'], ascending=False))
```

```python
df[['feature','corr_reviews']].set_index('feature').sort_values(by=['corr_reviews'], ascending=False)[:-1].plot.barh(legend=False)
plt.xlabel('Correlation')
plt.ylabel('')
plt.title('Correlation Between Rating Category and Number of Reviews')
plt.show()

value_calc.plot(kind='scatter',x='facilities',y='rating')
#plt.ylim(0,500)
```

```python
value_calc_nlp = value_calc.copy()
value_calc_nlp['description'] = value_calc_nlp['description'].fillna('')
value_calc_nlp['services'] = value_calc_nlp['services'].fillna('')
```

```python
value_calc_nlp['description'] = value_calc_nlp['description'].str.lower()
value_calc_nlp['services'] = value_calc_nlp['services'].str.lower()
```

```python
value_calc_nlp['description'] = value_calc_nlp['description'].apply(lambda x: re.sub('\s+', ' ', x))
value_calc_nlp['services'] = value_calc_nlp['services'].apply(lambda x: re.sub('\s+', ' ', x))
```

```python
# Filtering
value_calc_nlp['description'] = value_calc_nlp['description'].apply(lambda x: re.sub('[^\w\s]', '', x))
value_calc_nlp['services'] = value_calc_nlp['services'].apply(lambda x: re.sub('[^\w\s]', '', x))

stop = stopwords.words('english')
value_calc_nlp['description'] = value_calc_nlp['description'].apply(lambda text: " ".join(word for word in text.split() if word not in stop))
value_calc_nlp['services'] = value_calc_nlp['services'].apply(lambda text: " ".join(word for word in text.split() if word not in stop))
```

```python
# WordCloud
print('Vietnam')
wc = WordCloud(background_color="white", max_words=200, width=800, height=400)
wc.generate(' '.join(value_calc_nlp[value_calc_nlp['country']=='Vietnam']['name']))
plt.figure(figsize=(12, 6))
plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
plt.show()

print('Cambodia')
wc = WordCloud(background_color="white", max_words=200, width=800, height=400)
wc.generate(' '.join(value_calc_nlp[value_calc_nlp['country']=='Cambodia']['description']))
plt.figure(figsize=(12, 6))
plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
plt.show()

print('Laos')
wc = WordCloud(background_color="white", max_words=200, width=800, height=400)
wc.generate(' '.join(value_calc_nlp[value_calc_nlp['country']=='Laos']['description']))
plt.figure(figsize=(12, 6))
plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
# Sentiment Analysis
def sentiment_func(x):
    sentiment = TextBlob(x['description'])
    x['polarity'] = sentiment.polarity
    x['subjectivity'] = sentiment.subjectivity
    return x

value_calc_nlp = value_calc_nlp.apply(sentiment_func, axis=1)
```

```python
reviews = [1,2000]
rating = 1

value_calc_nlp[(value_calc_nlp['reviews'].between(reviews[0],reviews[1])) & (value_calc_nlp['rating']>rating) & (value_calc_nlp['country'].isin(trio))].plot.scatter('reviews', 'subjectivity')
plt.show()

value_calc_nlp[(value_calc_nlp['reviews'].between(reviews[0],reviews[1])) & (value_calc_nlp['rating']>rating) & (value_calc_nlp['country'].isin(trio))].plot.scatter('reviews', 'polarity')
plt.xlabel('Number of Reviews')
plt.ylabel('Polarity of Property Description')
plt.title('Effect of Property Description on Number of Reviews')
plt.show
```
