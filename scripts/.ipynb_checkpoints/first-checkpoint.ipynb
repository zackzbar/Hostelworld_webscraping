```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
plt.style.use('seaborn')

pd.set_option('display.max_rows', 999)
pd.set_option('display.min_rows', 999)
```

```python
full = pd.read_csv('full.csv', index_col=0)
```

```python
mains = ['name','kind','city','country','price','reviews','rating_cat','rating','value']
```

```python
full['country'] = full['country'].replace('Hong Kong China', 'Hong Kong')
```

```python
# COUNTING THE FULL DATASET
```

```python
full_countrycount = pd.DataFrame(full[full['reviews']>=20]['country'].value_counts())
full_countrycount
```

```python
full_countrycount.plot.pie(y='country',figsize=(10,10),legend=False,autopct='%1.0f%%')#,textprops={'color':"w"})
plt.show()
```

```python
# MEASURES OF CENTRALITY FOR EACH COUNTRY, VARIOUS COLUMNS
```

```python
price_med = full[full['reviews']>20].groupby('country')['price'].agg(['mean','median']).sort_values(by='median',ascending=False)
price_med
```

```python
price_med.plot.bar()
plt.ylabel('Price in USD')
plt.xlabel('')
plt.title('Measures of Price by Country')
plt.show()
```

```python
# VALUE CALCULATION AND RATING
```

```python
value_calc = full[(full['reviews']>=20) & (full['value']>0)].sort_values('value', ascending=False)
value_rate = full[(full['reviews']>=20) & (full['value']>0)].sort_values('value_for_money', ascending=False)
value_calc.shape
```

```python
display(value_calc[mains].head(100))
value_calc[mains].tail(10)
```

```python
mains.append('value_for_money')

display(value_rate[mains].head(100))
value_rate[mains].tail(10)
```

```python
value_top_count = pd.DataFrame(value_calc.head(250)['country'].value_counts())
#value_rate.head(500)['country'].value_counts()
display(value_top_count)
full_countrycount
```

```python
joined_count = pd.concat([value_top_count,full_countrycount], axis=1, join='inner')
joined_count.columns = ['value_count','total_count']
joined_count.index.name = 'country'
joined_count['value_percent'] = joined_count['value_count'] / sum(joined_count['value_count']) * 100
joined_count['total_percent'] = round(joined_count['total_count'] / 1418 * 100,1)
joined_count
```

```python
value_bottom_count = pd.DataFrame(value_calc.tail(250)['country'].value_counts())
#value_rate.tail(100)['country'].value_counts()
display(value_bottom_count)
full_countrycount
```

```python
joined_bottom_count = pd.concat([value_bottom_count,full_countrycount], axis=1, join='inner')
joined_bottom_count.columns = ['value_count','total_count']
joined_bottom_count.index.name = 'country'
joined_bottom_count['value_percent'] = joined_bottom_count['value_count'] / sum(joined_bottom_count['value_count']) * 100
joined_bottom_count['total_percent'] = round(joined_bottom_count['total_count'] / 1418 * 100,1)
joined_bottom_count
```

```python
joined_count[['value_percent','total_percent']].plot.bar()
plt.ylabel('Percent')
plt.xlabel('')
plt.title('% of Hostels in Top 20% for Value, vs. % of All Hostels')
plt.show()

joined_bottom_count.iloc[0:10,:][['value_percent','total_percent']].plot.bar()
plt.ylabel('Percent')
plt.xlabel('')
plt.title('% of Hostels in Bottom 20% for Value, vs. % of All Hostels')
plt.show()
```
