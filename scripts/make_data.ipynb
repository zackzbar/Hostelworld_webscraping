```python
# Quick look at the dimensions...
```

```python
countries = [thailand, myanmar, singapore, malaysia, vietnam, philippines, indonesia, laos, cambodia, brunei_timor, hong_kong]
names = ['thailand', 'myanmar', 'singapore', 'malaysia', 'vietnam', 'philippines', 'indonesia', 'laos', 'cambodia', 'brunei_timor', 'hong_kong']
zipped = list(zip(names, countries))
```

```python
for x, y in zipped :
    print(x,': \t',y.shape)
```

```python
# Concat into one big df, and take a look
```

```python
full = pd.concat(countries, axis=0, ignore_index=True)

cols = ['name', 'kind', 'city', 'country', 'description', 'services', 'price', 'distance', 'reviews', 'rating_cat', 'rating', 'atmosphere', 'cleanliness', 'location', 'security', 'facilities', 'staff', 'value_for_money']
full = full[cols]

display(full.head(3))
display(full.tail(3))
print(full.shape)
full.dtypes
```

```python
# Convert reviews to int, price to float, distance to float
```

```python
full['reviews'].fillna(0, inplace=True)
full['reviews'] = full['reviews'].apply(lambda x: int(x))
```

```python
full['distance'].mask(full['distance']=='unknown', inplace=True)
full['distance'] = full['distance'].apply(lambda x: float(x))
```

```python
# drops a fully null row I found...
full.drop(1011,axis=0,inplace=True)
```

```python
import re
full['price'] = full['price'].apply(lambda x: float(str(x)[3:]))
```
